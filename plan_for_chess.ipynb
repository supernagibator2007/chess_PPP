# Семинар 18.02 - создание заготовки плана и анализ требований

Требуется применить ООП. В Python оно реализуется через классы с назначением атрибутов и методов.

## Организация классов

Отдельно создать 'пустой' класс для фигур, который будет родительским для всех фигур. В нём будет метод по определению возможных ходов. Атрибутами будут цвет, координаты, символ для вывода в консоль.

Доска так же будет выступать отдельным классом. Объект 'доска' будет взаимодействовать с фигурами - с методом определения ходов (например, для осознания шаха/мата/пата), с координатами фигур (например, менять при ходе), а также будет отвечать за вывод в консоль. Атрибутами доски будут сами фигуры.

## Движение фигур

Первые ходят белые. Программа запрашивает ввод в консоль в следующем формате:

**С какого поля пойти:**



**На какое поле пойти:**

Если цвет фигуры правильный и ход возможный - программа выполняет ход и передаёт возможность второму игроку. Если нет - говорит, в чём проблема, и переспрашивает (реализация через while True). Кол-во ходов считается.

Если после хода возник шах/мат/пат, выводится соответствующее сообщение в консоль, а фигуры получают ограничения (вывести короля из-под удара/конец игры - определение победителя/ничья)
